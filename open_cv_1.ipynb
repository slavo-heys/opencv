```python
from google.colab.patches import cv2_imshow
import cv2

cv2.__version__
```

Wczytywanie obrazu

```python
img = cv2.imread(filename='view.jpg')
img
```

```python
cv2_imshow(img)
```

```python
img.shape
```

Zapisanie obrazu do pliku

```python
cv2.imwrite(filename='view-copy.png', img=img)
```

Wczytanie obrazu w odcieniu szaroci

```python
img = cv2.imread(filename='view.jpg', flags=cv2.IMREAD_GRAYSCALE)
img
```

```python
cv2_imshow(img)
```

```python
img.shape
```

Image Addition - dodawanie obrazw

```python
img1 = cv2.imread(filename='bear.jpg')
img2 = cv2.imread(filename='view.jpg')
```

```python
add_img = cv2.add(src1=img1, src2=img2)
add_img.shape
```

```python
cv2_imshow(add_img)
```
